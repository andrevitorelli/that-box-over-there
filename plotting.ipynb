## 1) Plotting contour regions around line plots

```python
lenx = 10
x = np.linspace(.1,10,lenx)

ylist = []

N = 100
for i in range(N):
  ylist += [np.random.normal(10/x,5/x)]
y=np.array(ylist)
```

```python
Ym, Y, Yp = np.array(list(map(lambda v: [v[0], v[1], v[2]], zip(*np.percentile(y, [16, 50, 84],axis=0))))).T
```

```python
def plotcontour(X,Y,Ym,Yp,label,color):
  plt.fill_between(X,np.array(Ym),
                   np.array(Yp),color=color,alpha=.1)
  plt.plot(X,Y,color=color,label=label)
```

```python
color='red'
label='experiment'
plt.style.use('ggplot')
plotcontour(x,Y,Ym,Yp,label,color)
plt.yscale('log')
```
