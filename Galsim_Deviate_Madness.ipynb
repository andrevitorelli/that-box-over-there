```python
seed = 31415
base = galsim.BaseDeviate(seed)
im = galsim.InterpolatedImage(galsim.Image(np.ones((2,2)),scale=1.)).drawImage(nx=2,ny=2)
for i in range(20):
  noise = galsim.CCDNoise(base.duplicate())
  print (noise._rng)
  im.addNoise(noise)

  
```
