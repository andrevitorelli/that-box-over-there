```python
from matplotlib import patches
```

```python

fig  = figure()
ax  = fig.add_subplot(111)
xcenter = 0
ycenter = 0
width=1
height=2
angle=30
for i in range(10):
    e2 = patches.Ellipse((xcenter, ycenter), width*i, height*i,
                     angle=angle, linewidth=0, color='gray', fill=True, zorder=2,alpha=.1)
    ax.add_patch(e2)
xlim(-10,10)
ylim(-10,10)
xlabel("$\\theta$",fontsize=20)
ylabel("$x$",fontsize=20)
axvline(2.5,color='darkblue',label="$p(x|\\theta)$")
axhline(2.5,color='darkred',label="$p(\\theta|x)$")
ax.set_xticks([])
ax.set_yticks([])
legend(fontsize=20)
title("$p(x,\\theta)$",fontsize=20)
```
